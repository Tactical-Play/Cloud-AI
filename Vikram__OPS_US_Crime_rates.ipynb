### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
crime.index
```

```python
def dec(x):#creating a function to group years to decades
    x=str(x)
    if x>='1960-01-01' and x <'1970-01-01':
        return '1960-01-01'
    elif x <'1980-01-01':
        return '1970-01-01'
    elif x <'1990-01-01':
        return '1980-01-01'
    elif x <'2000-01-01':
        return '1990-01-01'
    else: return '2000-01-01'
crime_dec=crime.groupby(dec).sum()
```

```python
crime_dec# population has been summed, needs to be rectified
```

```python
sum(crime.loc['1960-01-01':'1969-01-01','Violent'])# just checking
```

```python
population_mean = crime['Population'].groupby(dec).mean()# take MEAN value for Population column
population_mean = pd.DataFrame(population_mean.astype(int))
```

```python
crime_dec['Population']=population_mean
crime_dec # finally replaced population by its mean for each decade
```

```python
#Alternate Way - take max of population
crime_alt = crime_dec.copy()
population_max = crime['Population'].groupby(dec).max()# take MAX value for Population column
population_max = pd.DataFrame(population_max.astype(int))
```

```python
crime_alt['Population']=population_max
crime_alt # finally replaced population by its max for each decade
```

### Step 9. What is the most dangerous decade to live in the US?
